## 2. Building first ML model

```
#Import Libraries
from sklearn.tree import DecisionTreeRegressor
```

## 3. Model Validation

> How trustable is that model? Is it correctly used?

The goal of model validation is to measure the accuracy of it (This is viable to apply in MOST of cases)

### Metrics

There are metrics that help us with this task

- Mean Absolute Error (MAE)

> As we are comparing almost the same data the only thing that makes the predicted model different in the quantity of features or parameters in X

```
from sklearn.metrics import mean_absolute_error

predicted_home_prices = melb_model.predict(X) 

print('Mean Absolute Error')
print(mean_absolute_error(Y, predicted_home_prices)) #Comparing the price in function of having the whole info
print('--------------------------------------------')

no_match = 0
for i,j,z in zip(melb_data['Price'], predicted_home_prices, range(len(predicted_home_prices))):
    
    if i != j:
        no_match += 1
        print('Row', z+1, 'Original Value', i, 'Predicted Value', j)

print("Total values that don't match", no_match)        
```

### The problem of accuracy

There could be a problem, as we analyzed the whole dataset (with just excluding some parameters) and then compared to the same data, the MAE doesn't show so much problem or inaccuracy, right? Well, we haven't worked with so new data, and that is a big problem, because imagine, that the data we trained or fit the model with, maybe concludes that having a green door increases the price of the house, then, with new data we could have an accuracy problem

> IT'S TOTALLY IMPORTANT TO TEST THE MODEL WITH NEW INFORMATION, AND CALCULATE THE METRICS, WE CANNOT GIVE A CONCLUSION COMPARING IT WITH THE NEARLY THE SAME INFO

Let's analyze again but splitting the data into 2, the training data and the 'new' data we will pass to predict the price

```
from sklearn.model_selection import train_test_split #This library helps us to split the info randomly

train_x, new_data_x, train_y, data_y = train_test_split(X, Y, random_state=0) #remember random_state is like the .seed

melb_model_2 = DecisionTreeRegressor(random_state=0)
melb_model_2.fit(train_x, train_y)

predicted_home_prices_2 = melb_model_2.predict(new_data_x)
print('New MAE Calculated')
mean_absolute_error(data_y, predicted_home_prices_2) #measures accuracy with real values
```

## 4. Underfitting and Overfitting

> When we divide the houses amongst many leaves, we also have fewer houses in each leaf. Leaves with very few houses will make predictions that are quite close to those homes' actual values, but they may make very unreliable predictions for new data (because each prediction is based on only a few houses).                                                                                                                                                                                        This is a phenomenon called overfitting, where a model matches the training data almost perfectly, but does poorly in validation and other new data. On the flip side, if we make our tree very shallow, it doesn't divide up the houses into very distinct groups. 

> At an extreme, if a tree divides houses into only 2 or 4, each group still has a wide variety of houses. Resulting predictions may be far off for most houses, even in the training data (and it will be bad in validation too for the same reason). When a model fails to capture important distinctions and patterns in the data, so it performs poorly even in training data, that is called underfitting.

<img src="image-20240412-180959.png" width="" align="" />

So then, the middle point of this graph is the one we would want to get to and between red and blue line, so the problem to solve is to control the tree depth

We'll use a function to 

```
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000, 10000]:
    my_mae = get_mae(max_leaf_nodes, train_x, new_data_x, train_y, data_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

> Summary, Overfitting: capturing spurious(syn: False) patterns that won't recur in the future, leading to less accurate predictions, or
Underfitting: failing to capture relevant patterns, again leading to less accurate predictions.

## 5. Random Forests

> The random forest uses many trees, and it makes a prediction by averaging the predictions of each component tree. It generally has much better predictive accuracy than a single decision tree and it works well with default parameters. If you keep modeling, you can learn more models with even better performance, but many of those are sensitive to getting the right parameters.

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error #already imported


random_forest_model = RandomForestRegressor(random_state=1)
random_forest_model.fit(train_x, train_y)
prediction_rand_forest = random_forest_model.predict(new_data_x)
print(mean_absolute_error(data_y, prediction_rand_forest))
```
